```python
import pdfplumber
import pandas as pd

all_tables = []
with pdfplumber.open(r"C:\Users\Asheem Siwach\Downloads\List of PAN India APF Projects.pdf") as pdf:
    for pageno, page in enumerate(pdf.pages, start=1):
        # extract_tables works best for clearly delimited text columns
        tables = page.extract_tables()
        all_tables.extend(tables)

print(all_tables)
```

```python
import pandas as pd

dfs = []
for i ,table in enumerate(all_tables):
    print(f"Table {i+1}:")
    table = pd.DataFrame(table, columns=["State","City","Builder Group","Project Name","Project Address"])
    dfs.append(table)
```

```python
all_dfs = pd.concat(dfs)
all_dfs = all_dfs.iloc[1:,:]

# Clean the data and handle None values
for column in all_dfs.columns:
    all_dfs[column] = all_dfs[column].astype(str).str.strip()
    all_dfs[column] = all_dfs[column].str.capitalize()

# Remove rows with None or empty values
all_dfs = all_dfs[all_dfs['State'] != 'None']
all_dfs = all_dfs[all_dfs['State'] != '']

dictionary = {}

for state in all_dfs['State'].unique():
    if pd.isna(state) or state == 'None':
        continue
    state_df = all_dfs[all_dfs['State'] == state]
    dictionary[state] = {}
    
    for city in state_df['City'].unique():
        if pd.isna(city) or city == 'None':
            continue
        city_df = state_df[state_df['City'] == city]
        dictionary[state][city] = {}
        
        for builder_group in city_df['Builder Group'].unique():
            if pd.isna(builder_group) or builder_group == 'None':
                continue
            builder_group_df = city_df[city_df['Builder Group'] == builder_group]
            dictionary[state][city][builder_group] = {}
            
            for index, row in builder_group_df.iterrows():
                project_name = row['Project Name']
                # Use project name as key, or index if project name is empty/None
                project_key = project_name if project_name and project_name != 'None' else f"Project_{index}"
                dictionary[state][city][builder_group][project_key] = {
                    "Project Name": project_name,
                    "Project Address": row['Project Address']
                }

print(f"Dictionary created with {len(dictionary)} states")
print("Sample structure:", list(dictionary.keys())[:3] if dictionary else "No data")
```
